### Tokenization:-

* Tokenization is the process of dividing a sequence of text, such as a sentence or a paragraph, into individual units called tokens. These tokens can be words, characters, or even smaller units like subwords or syllables, depending on the specific tokenization approach used.



* Tokenization is a crucial step in natural language processing (NLP) and text analysis because it breaks down text into smaller meaningful units, enabling further analysis and processing. By splitting text into tokens, we can analyze and manipulate the text on a granular level, which is necessary for tasks like sentiment analysis, text classification, machine translation, and many others.




* For example, consider the sentence: "The cat is sitting on the mat." After tokenization, this sentence could be represented as a list of tokens: ["The", "cat", "is", "sitting", "on", "the", "mat", "."]. Each word and the punctuation mark are considered as separate tokens.




* Tokenization can involve several considerations, such as handling punctuation, dealing with contractions (e.g., splitting "can't" into "can" and "'t"), recognizing special tokens (e.g., numbers or URLs), and more. Different tokenization techniques and libraries, like NLTK or spaCy in Python, provide various approaches to tokenization, allowing you to choose the one that best suits your specific NLP task.

### Stemming

* Stemming is a process used in natural language processing (NLP) to reduce words to their base or root form, known as a stem. It involves removing prefixes, suffixes, and sometimes even parts of the word's interior to obtain the root form.

* Let's simplify the concept of stemming with a layman's example:

* Imagine you have a collection of words related to the concept of "running": run, runs, running, runner, and ran. The stem of all these words is "run" because they share the same basic meaning.

* Stemming algorithms apply specific rules to identify and remove common word affixes. By doing so, they convert related words into their common base form. This process allows us to consolidate different variations of a word into a single representation, which can be useful for various NLP tasks.

* Here are a few examples to illustrate stemming:

* 1. Word: "Jumping"
 *   Stem: "Jump"

* 2. Word: "Happiness"
 *   Stem: "Happi"

* 3. Word: "Dancing"
*    Stem: "Danc"

* As you can see, the stemming process reduces words to their core or base form, often by removing endings like "-ing" or "-ness". This helps to capture the underlying meaning of words without being concerned about grammatical variations.

* Stemming is used in applications like search engines, text mining, and sentiment analysis, where understanding the general concept of a word is more important than its specific grammatical form. **However, it's important to note that stemming is a crude approximation, and it may not always produce accurate stems or preserve the exact meaning of words. There can be cases where the resulting stem may not be a valid word in the language**.

* There are different stemming algorithms available, such as the **Porter stemming algorithm, Snowball stemming algorithm, and Lancaster stemming algorithm**, each with its own rules and effectiveness depending on the language and context of the text.

### Lemmatisation

* Lemmatisation, similar to stemming, is a process used in natural language processing (NLP) to reduce words to their base form. However, unlike stemming, **lemmatisation takes into account the word's meaning and context, aiming to produce valid words known as lemmas**.

* Let's explain lemmatisation with a layman's example:

* Consider a sentence with the words "am," "are," and "is." The lemma for all these words is "be" because they are different forms of the verb "to be." Lemmatisation recognizes that these variations share the same underlying meaning.

* Lemmatisation relies on linguistic knowledge and uses dictionaries or language resources to identify the correct base or lemma form of a word. It takes into account factors such as part of speech (noun, verb, adjective, etc.) and inflectional endings to produce meaningful lemmas.

* Here are a few examples to illustrate lemmatisation:

* 1. Word: "Running"
*    Lemma: "Run"

* 2. Word: "Cats"
*   Lemma: "Cat"

* 3. Word: "Better"
*   Lemma: "Good"

* As you can see, lemmatisation provides more accurate base forms compared to stemming because it considers the word's context and meaning. It handles irregularities and exceptions in a language, producing valid words that retain their semantic integrity.

* Lemmatisation is useful in NLP tasks where maintaining the exact meaning and grammatical correctness of words is crucial. It is commonly used in applications like machine translation, question-answering systems, and sentiment analysis, where understanding the precise word form is important for accurate analysis and interpretation.

* It's worth noting that lemmatisation can be computationally more intensive than stemming because it requires more linguistic knowledge and resources. However, it often produces more accurate results by preserving the semantics and grammatical structure of words.

### Stopwords

* In natural language processing (NLP), stopwords are common words that are considered insignificant or have little contribution to the overall meaning of a sentence or text. These words are often filtered out or removed from the text during preprocessing to improve the efficiency and focus of NLP tasks.

* Stopwords typically include words like "a," "an," "the," "in," "is," "and," "of," and other frequently occurring words that do not carry significant meaning or contribute much to the analysis or understanding of the text.

* Here's an example to help illustrate stopwords:

* Consider the sentence: "The cat is sitting on the mat."

* In this sentence, the stopwords are "the" and "is." These words are commonly used in English but don't provide much specific information about the cat or the action of sitting. Removing these stopwords would result in the sentence: "cat sitting on mat."

* By removing stopwords, we can focus more on the meaningful words in the text, such as "cat" and "sitting," which convey more specific and relevant information.

* The use of stopwords removal can benefit various NLP tasks, including text classification, information retrieval, sentiment analysis, and topic modeling. By eliminating these common words, the analysis can be more focused on the essential content words, improving the efficiency and accuracy of the NLP algorithms.

* It's important to note that the list of stopwords can vary depending on the specific language, context, or application. Different libraries and NLP frameworks provide predefined lists of stopwords, but you can also customize the list based on your specific needs or domain-specific stopwords.

* Overall, stopwords are common words that are removed from text during NLP preprocessing to improve the efficiency and relevance of subsequent NLP tasks by eliminating words that carry little meaning or significance.

### Bag of Words

* The "bag of words" is a technique used in natural language processing to represent text data. It involves counting the occurrences of words in a document without considering the order or structure of the words.

* Imagine you have a collection of sentences: 

* 1. "The cat is on the mat."
* 2. "The dog is on the mat."
* 3. "The cat and the dog are friends."

* To create a bag of words representation, you would first make a list of all the unique words in these sentences:

* ["The", "cat", "is", "on", "the", "mat", "dog", "and", "are", "friends"]

* Next, you would count the number of times each word appears in each sentence and create a vector for each sentence. The order of the words doesn't matter. Let's represent each sentence as a vector:

* 1. [2, 1, 1, 1, 1, 1, 0, 0, 0, 0]
* 2. [2, 0, 1, 1, 1, 1, 1, 0, 0, 0]
* 3. [2, 1, 0, 0, 1, 0, 1, 1, 1, 1]

* In each vector, the number at each position corresponds to the count of the word at that position in the word list.

* This bag of words representation is useful for various tasks like sentiment analysis, text classification, or information retrieval. Although it loses information about word order, it simplifies the text to numerical features that can be used by machine learning algorithms to analyze or classify the text.

* The disadvantages of Bag of words is that if the frequency of two words is same then it will give the same weight to both of words,Even if the second word has most important for the sentiment which we analyse but bag words will give equal weightage so **thats why bag words not prefer much instead of it we use TF-IDF(Term Frequecy Inverse Document Frequency)**

### TF-IDF(Term Frequency-Inverse Document Frequency)

* Certainly! TF-IDF (Term Frequency-Inverse Document Frequency) is a technique used to evaluate the **importance of words in a document or a collection of documents**. It combines two factors: term frequency and inverse document frequency.

* **1. Term Frequency (TF)**: This measures how frequently a word appears in a document. It assumes that the more times a word appears, the more important it is in that particular document.
* **Formula of TF= (Number of times the word present in that sentence)/Total number of words present in that sentence**

* 2. **Inverse Document Frequency (IDF)**: This measures how rare or unique a word is across multiple documents. It assumes that words that are less common across documents are more informative and carry more significance.

* **Formula for IDF=log[(Total number of sentence)/(Number of sentences that conataning that word)]**
* To understand TF-IDF, let's consider an example. Imagine we have a collection of documents about animals, and we want to analyze the importance of words in a document titled **"The Cat's Life"**:

* **"The cat is a small, independent creature that is known for its agility and hunting skills. Cats are popular pets and are often found in households. The cat's fur can vary in color and length."**

* First, we calculate the Term Frequency (TF) for each word in the document. We count how many times each word appears:

- "the": 2
- "cat": 3
- "is": 1
- "a": 1
- "small": 1

- "creature": 1
- "that": 1

- "color": 1

* Next, we calculate the Inverse Document Frequency (IDF) for each word. We determine how often each word appears in the entire collection of documents. **For simplicity, let's assume we have a total of 100 documents, and the word "cat" appears in 10 of them. The IDF for "cat" would be**:

* **IDF("cat") = log(100 / 10) = log(10) = 1**

* Finally, we calculate the TF-IDF score for each word in the document by multiplying the Term Frequency (TF) by the Inverse Document Frequency (IDF). This score reflects the importance of each word within the document and across the collection:

- "the": 2 * IDF("the")

### Word Embedding:- 

* **Word embeddings is a technique where individual words are transformed into a numerical representation of the word (a vector).**
* Where each word is mapped to one vector, this vector is then learned in a way which resembles a neural network. The vectors try to capture various characteristics of that word with regard to the overall text. These characteristics can include the semantic relationship of the word, definitions, context, etc. With these numerical representations, you can do many things like identify similarity or dissimilarity between words.

* **Word embedding is a way of representing words as vectors. The main goal of word embedding is to convert the high dimensional feature space of words into low dimensional feature vectors by preserving the contextual similarity in the corpus**.

* Clearly, these are integral as inputs to various aspects of machine learning. A machine cannot process text in its raw form, thus converting the text into an embedding will allow users to feed the embedding to classic machine learning models. The simplest embedding would be a one hot encoding of text data where each vector would be mapped to a category.

### Word To Vec(Word to Vector):-

* Word2Vec is a popular technique in natural language processing (NLP) **used to represent words as numerical vectors**. It captures the **semantic meaning and relationships between words by mapping them to high-dimensional vector spaces**.

* In simple terms, Word2Vec learns to represent words in such a way that **similar words have similar vector representations**. It does this by analyzing large amounts of text data and learning patterns from the surrounding words.

* There are two main approaches in Word2Vec: Continuous Bag of Words (CBOW) and Skip-gram.

1. Continuous Bag of Words (CBOW):
   - In CBOW, the model predicts the target word based on the context words around it.
   - For example, given the sentence: "The cat sat on the mat," the context words for the target word "sat" might be "The," "cat," "on," and "the."
   - The model learns to predict the target word "sat" based on these context words.

2. Skip-gram:
   - In Skip-gram, the model predicts the context words given a target word.
   - Using the same example sentence, the model would predict the context words "The," "cat," "on," and "the" based on the target word "sat."
   - The model learns to generate the likely context words based on the target word.

* Both CBOW and Skip-gram models use neural networks to learn the word representations. The resulting word vectors capture the semantic relationships between words. For instance, words with similar meanings or that appear in similar contexts will have similar vector representations.

* These word vectors can be used in various NLP tasks, such as word similarity calculations, text classification, sentiment analysis, or even generating word analogies (e.g., king - man + woman = queen).

* Word2Vec has revolutionized NLP by providing a way to represent words as vectors and capture the rich semantic meaning of words. It allows machines to understand and work with human language in a more meaningful way.

* **Word2Vec model is used for Word representations in Vector Space which is founded by Tomas Mikolov and a group of the research teams from Google in 2013. It is a neural network model that attempts to explain the word embeddings based on a text corpus.**
